### Списки

```python
# Списки хранят последовательности 
li = [] # 

# Можно сразу начать с заполненного списка 
other_li = [4, 5, 6]

# Объекты добавляются в конец списка методом append() 
li.append(1) # [1] 
li.append(2) # [1, 2] 
li.append(4) # [1, 2, 4] 
li.append(3) # [1, 2, 4, 3] 
print(li, len(li))

# И удаляются с конца методом pop() 
print(li.pop()) # => возвращает 3 и li становится равен [1, 2, 4]
print(li)
li.append(3)

# Обращайтесь со списком, как с обычным массивом 
print( li[0] ) # => 1 

#Обратимся к последнему элементу 
print( li[-1] ) # => 3
```

```python
# Можно обращаться к диапазону, используя так называемые срезы 
print(li)
print(li[1:3]) # Вернуть список из индекса с 1 по 3 => [2, 4] 
print(li[2:]) # Вернуть список, начиная с индекса 2 => [4, 3] 
print(li[:3]) # Вернуть список с начала до индекса 3 => [1, 2, 4] 
print(li[::2]) # Вернуть список, выбирая каждую вторую запись => [1, 4] 
print(li[::-1]) # Вернуть список в обратном порядке => [3, 4, 2, 1] 
```

```python
# Используйте сочетания всего вышеназванного для выделения более сложных срезов 
# li[начало:конец:шаг]

print(li[1:4:2])
```

### Словари

```python
# Словари содержат ассоциативные массивы 
empty_dict = {} 

# Вот так описывается предзаполненный словарь 
filled_dict = {"one": 1, "two": 2, "three": 3}

# Обратите внимание, что ключи для словарей должны быть неизменяемыми типами. Это сделано для того, чтобы ключ может быть преобразован в хеш для быстрого поиска.  Неизменяемые типы включают целые числа, числа с плавающей запятой, строки, кортежи. 
#invalid_dict = {[1,2,3]: "123"} # => Выбрасывает ошибку TypeError: unhashable type: 'list' 

valid_dict = {(1,2,3):[1,2,3]} # Однако значения могут быть любого типа. 

# Поиск значений с помощью [] 
print(filled_dict["one"]) # => 1

# Все ключи в виде списка получаются с помощью метода keys().
print( list(filled_dict.keys()) ) # => ["one", "two", "three"]
```

### Лямбда-функции

https://webcache.googleusercontent.com/search?q=cache:Upz9HYtTFPQJ:https://habr.com/ru/company/piter/blog/674234/&cd=1&hl=ru&ct=clnk&gl=ru

- Эта функция может иметь любое количество аргументов, но вычисляет и возвращает только одно значение
- Лямбда-функции применимы везде, где требуются объекты-функции 
- Вы должны помнить, что  синтаксически лямбда-функция ограничена, позволяет представить всего одно выражение
- Они имеют множество вариантов применения в конкретных областях программирования, наряду  с другими типами выражений, используемых в функциях.

Лямбда-функции используют вместе с такими встроенными функциями как filter(), map(),reduce() и др.

```python
double = lambda x: x*2
print(double(5))

'''
Эквивалентна:

def double(x):
    return x * 2

'''
```

```python
my_list = [1, 3, 4, 6, 10, 11, 15, 12, 14]
new_list = list(filter(lambda x: (x%2 == 0) , my_list))
print(new_list)
```

```python
current_list = [1, 3, 4, 6, 10, 11, 15, 12, 14]
new_list = list(map(lambda x: x*2 , current_list))
print(new_list)
```

### \*args, \**kwargs

https://towardsdatascience.com/args-kwargs-python-d9c71b220970

```python
def sum_nums(*args):
    sum = 0
    for n in args:
        sum += n
    return sum

print(sum_nums(10, 20))
print(sum_nums(10, 20, 30))
print(sum_nums(5))
```

```python
def my_func(**kwargs):
    for key, val in kwargs.items():
        print(key, val)
        
my_func(a='hello', b=10)
my_func(param1=True, param2=10.5)
```

## Классы

```python
import random

class Animal():
    def __init__(self, name, size, color):
        self.name = name
        self.size = size
        self.color = color
        self.legs = 4
        self.position = [0, 0]
    
    def __call__(self, a, b):
        print(a * b)

    def new_Location(self):
        self.position = [random.uniform(0,10), random.uniform(0,10)]
    
    def say_Hi(self):
        return "Hello! I'am " + self.name   
```

```python
anim1 = Animal("Dog", 15, "black")
anim2 = Animal("Cat", 10, "white")
anim3 = Animal("Frog", 2, "green")

anim3(4,5)
print(anim3.say_Hi())

print(anim3.position)
anim3.new_Location()
print(anim3.position)
```

### Базовые операции над массивами

```python
a = np.array(([[0, 1, 2], [10, 12, 13]], [[100, 101, 102], [110, 112, 113]]))
print(a)

x = a.shape
print('Форма массива = ', x)

# Альтернативные способы получения среза массива
print(a[1, : , :])
print(a[1])
print(a[1, ...])

# Альтернативные способы получения среза массива
print(a[: , : , 2])
print(a[..., 2])

b = a.flat
print(b[8])

a.ravel()

a.reshape(6, 2)
```

```python
a = np.array([[1, 2, 3], [4, 5, 6]])
print(a)

x = np.sum(a)
print('Сумма всех элементов массива = ', x)

x = a.sum()
print('Сумма всех элементов массива = ', x)

x = a.min()
print('Минимальный элемент массива = ', x)

x = a.max()
print('Максимальный элемент массива = ', x)

x = a.min(axis=0)
print('Минимальный элемент в каждом столбце = ', x)

x = a.min(axis=1)
print('Минимальный элемент в каждом каждой строке = ', x)

x = a.argmax()
print('Индекс максимального элемента = ', x)

x = a.argmin()
print('Индекс минимального элемента = ', x)
```

```python
a = np.random.randint(100, size=(5, 4))
print('Исходный массив:')
print(a)

print(a[:, 0])
print(a[:, 0].argsort())

a = a[a[:, 0].argsort()]
print('Отсортированный по первому столбцу массив:')
print(a)

a = a[:, a[0, :].argsort()]
print('Отсортированный по первой строке массив:')
print(a)
```

```python
def f(x, deriv=False): 
    if deriv:
        return (f(x+.001) - f(x))/.001        
    return 1/(1 + np.exp(-x))

f(10,True)

x = np.linspace(-10,10,100)

x1 = 2
y1 = f(x1)

def line(x, x1, y1):
    return f(x1,True)*(x - x1) + y1

xrange = np.linspace(x1-3, x1+3, 10)

fig, ax = plt.subplots()
ax.grid(which="major", linewidth=1.2)
ax.grid(which="minor", linestyle="--", color="gray", linewidth=0.5)
plt.plot(x, f(x))
plt.scatter(x1, y1, color='C1', s=50)
plt.plot(xrange, line(xrange, x1, y1), 'C1--', linewidth = 2)
plt.show()
```
